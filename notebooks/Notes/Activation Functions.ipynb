```python
import ipywidgets
import numpy as np
import tensorflow as tf
import matplotlib.pyplot
%matplotlib inline

sess = tf.Session()

def init_plot():
    plt = matplotlib.pyplot
    plt.figure(figsize=(9, 9), dpi=80, facecolor='w', edgecolor='k')
    plt.ylim(-1.0, 4.0)
    plt.grid(True, which='both', color='lightblue', linestyle="dotted")
    plt.axhline(y=0, color='lightgrey', linestyle="dashed")
    plt.axvline(x=0, color='lightgrey', linestyle="dashed")
    return plt
```

# Activation Functions

### Softmax

```python
x = tf.Variable([2.0, 2.0, 2.0, 4.0], 'x')
r = tf.nn.softmax(x)
sess.run(tf.global_variables_initializer())
sess.run(r)
```

### Rectified Linear Unit (ReLU)

```python
x = tf.Variable([2.0, -20.0, -2.0, 4.0], 'x')
r = tf.nn.relu(x)
sess.run(tf.global_variables_initializer())
sess.run(r)
```

### Logistic Function

Also referred to as the sigmoid function. tf.sigmoid is exactly $ y = 1 / (1 + exp(-x)) $ 

```python
x = np.linspace(-6, 6, 11)
p = init_plot()
p.ylim(-0.2, 1.2)

y = 1 / (1 + np.e ** -x)
p.plot(x, y, color="blue")

r = tf.sigmoid(x)
sess.run(tf.global_variables_initializer())
y = sess.run(r)
p.plot(x, y, linestyle="dashed", color="lightblue")
```

### tanh()

```python
x = np.linspace(-6, 6, 44)
p = init_plot()
p.ylim(-2.2, 2.2)

y = np.tanh(x) * 0.5 + 0.5
p.plot(x, y, color="blue")

y = np.tanh(x)
p.plot(x, y, color="red")
```
