```python
import ipywidgets
import numpy as np
import matplotlib.pyplot
%matplotlib inline

def scatter_plot():
    plt = matplotlib.pyplot
    size = 9
    golden_ratio = 1.618
    plt.figure(figsize=(golden_ratio * size, size), facecolor='w', edgecolor='k')
    plt.grid(True, which='both', color='lightblue', linestyle="dotted")
    plt.axhline(y=0, color='lightgrey', linestyle="dashed")
    plt.axvline(x=0, color='lightgrey', linestyle="dashed")
    plt.ylim(-0.2, 1.2)
    return plt

def histogram_plot(x):
    plt = matplotlib.pyplot
    n, bins, patches = plt.hist(x, 128, facecolor='g', alpha=0.75)

xd = np.linspace(0.0, 1.0, 1024)
```

## Pseudo Randomness

$$
f_{prnd}(x) = frac(f(x * n))
$$

Where $f(x)$ returns values ranging from $0 <= y < 1$ for all values of $x$. Let $n$ be very large. For a large $n$ the result is seemingly random since the value of the fractional part of $x * n$ is not smooth over $0 <= x < 1$, and is further obfuscated by function $f(x)$

```python
def frac(x):
    return np.mod(x, 1.0)

# f(x) functions
def f_pow(x):
    return pow(x, 2.0)

def f_sin(x):
    return frac(np.sin(x * 2.0 * np.pi))

# Pseudo-random
def f_frac_mult(x):
    return frac(x * 125357.77)

def f_pow_mult(x):
    return frac(f_pow(x * 11267.0))

def f_sin_mult(x):
    return frac(f_sin(x * 1586.0))

def f_combined(x):
    return f_frac_mult(f_sin_mult(x));

# For comparison
def f_np_rand(x):
    return np.random.rand(len(x))
```

Distributions from these functions...

```python
y_sin = f_sin(xd)
histogram_plot(y_sin)
```

```python
y_pow = f_pow(xd)
histogram_plot(y_pow)
```

```python
y_frac_mult = f_frac_mult(xd)
histogram_plot(y_frac_mult)
```

```python
y_pow_mult = f_pow_mult(xd)
histogram_plot(y_pow_mult)
```

```python
y_np_rand = f_np_rand(xd)
histogram_plot(y_np_rand)
```

```python
y_sin_mult = f_sin_mult(xd)
histogram_plot(y_sin_mult)
```

```python
y_combined = f_combined(xd)
histogram_plot(y_combined)
```
