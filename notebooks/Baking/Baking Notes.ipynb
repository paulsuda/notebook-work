
# Pie Crust

Makes double normal recipe.

2 1/4 cups white flour
1/8 cups sugar
1/2 tsp salt
1 1/4 cup butter (or substitute)
1/2 cup ice water

1. Blend flour, salt, and sugar in food processor

2. Add cubed chilled butter, blend until butter is flaked.

3. Drop in cold water 1 tablespoon at a time. Pulse food processor until dough ball forms.

4. Wrap in plastic and refrigerate for 4 or more hours.

5. Roll dough to fit 9 inch pie pan, press evenly into sides and bottom of pan. Use extra flour on board and rolling pin. Keep dough cool while working.

## Hints:

* Work with ingredients as cold as possible. Including chilling the flour and the food processor bowl.

* Partially substitute butter with 1/2 cup shortening or Crisco.

From: https://www.allrecipes.com/recipe/24094/butter-flaky-pie-crust/

# Pumpkin Pie

Preheat over to 425F.

1 pie crust
whipped cream (optional)

+ Pie Filling:
1/2 tsp salt
1 tsp ground cinnamon
1/2 tsp ground ginger
1/2 tsp ground nutmeg (optional)
1/4 tsp ground cloves (optional)
2 eggs
1 15oz can pumpkin

 - one of -

3/4 cup granulated sugar (or substitute)
1 12oz can evaporated milk

 - or -

1 14oz can sweetened condensed milk

 - choose evaporated or sweetened condensed milk -

1. Preheat oven.

2. Mix all filling ingredients with electric hand mixer.

3. Pour filling into pie crust.

4. Bake 15  minutes at 425F.

5. Reduce temp and bake 35 to 40 minutes at 350F. Knife should come out clean when done.

6. Remove from oven and cool. Serve with whipped cream.

Hint: Substitute brown or dark brown sugar for 1/4 or 1/2 of granulated.

From: https://www.verybestbaking.com/recipes/18470/LIBBYS-Famous-Pumpkin-Pie
From: https://www.eaglebrand.com/recipe-detail/traditionalpumpkinpie-4117

# Amaretti Cookie

+ Dough
2 cups almond flour
1/2 cup granulated sugar
1/2 cup confectioner sugar
1/4 teaspoon salt
2 large egg whites
1/2 teaspoon almond extract

+ Coating
additional confectioners sugar for rolling


1. Combine and mix dry ingredients in mixing bowl.

2. Add egg whites and extract to dry ingredients, mix until dough.

3. Wrap dough ball in plastic and let rest for 1 hour.

4. Preheat over to 325 F, line sheet with parchment.

5. Form 1" diameter dough balls, roll in powdered sugar.

6. Place 1 1/2" apart on sheet, bake for 25 to 30 min.

7. Remove cookies, allow to cool, take off sheet.

From: https://www.kingarthurflour.com/recipes/amaretti-recipe
